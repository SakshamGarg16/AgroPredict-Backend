```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pickle
```

```python
data = pd.read_csv(r"../dataset/Fertilizer Prediction.csv")
data
```

```python
soil_ty = LabelEncoder()
Crop_ty = LabelEncoder()
Ferti_nam = LabelEncoder()

data['Soil_Type'] = soil_ty.fit_transform(data['Soil Type'])
data['Crop_Type'] = Crop_ty.fit_transform(data['Crop Type'])
data['Fertilizer_Name']= Ferti_nam.fit_transform(data['Fertilizer Name'])
data
```

```python
Ferti_nam.classes_
```

```python
inputs = data.drop(['Soil Type','Crop Type','Fertilizer Name','Fertilizer_Name'],axis='columns')
inputs
```

```python
target = data.drop(['Temparature','Humidity ','Moisture','Nitrogen','Potassium','Phosphorous',	'Soil_Type' , 'Crop_Type','Soil Type','Crop Type','Fertilizer Name'],axis = 'columns')
target
```

```python

tempmed = inputs.Temparature.median()
Humid = inputs['Humidity '].median()
Moist = inputs.Moisture.median()
Nitro=inputs.Nitrogen.median()
Pota = inputs.Potassium.mean()
Phos = inputs.Phosphorous.median()

print(Humid,Moist,Nitro,Pota,Phos,tempmed)
```

```python
inputs
```

```python

inputs.Temparature = inputs.Temparature.fillna(tempmed)
inputs['Humidity '] = inputs['Humidity '].fillna(Humid)

inputs.Moisture = inputs.Moisture.fillna(Moist)
inputs.Nitrogen = inputs.Nitrogen.fillna(Nitro)

inputs.Potassium = inputs.Potassium.fillna(Pota)
inputs.Phosphorous = inputs.Phosphorous.fillna(Phos)

inputs
X_train,X_Test,Y_train,Y_Test= train_test_split(inputs,target,test_size=0.2,shuffle=True)
X_train
```
