```python
import pandas as pd
# import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
import pickle
# from sklearn import neural_network
```

```python
data2 = pd.read_csv(r"../dataset\PRICE_DATASET.csv")
data2
```

```python

encode = LabelEncoder()

data=data2
data['STATE'] = encode.fit_transform(data['STATE'])
data['DISTRICT'] = encode.fit_transform(data['DISTRICT'])
data['MARKET'] = encode.fit_transform(data['MARKET'])
data['COMMODITY']=encode.fit_transform(data['COMMODITY'])
data['VARITY'] = encode.fit_transform(data['VARITY'])

data.head(10)
```

```python
data = data.drop(['arrival_date'],axis='columns')
inputs = data.drop(['modal_price'],axis='columns')
inputs
```

```python
target = data['modal_price']
target
```

```python
X_Train,X_Test,Y_Train,Y_Test = train_test_split(inputs,target,train_size=0.9)
```

```python
# model5=linear_model.LogisticRegressionCV(max_iter=1000)
# model5.fit(X_Train,Y_Train)
# model5.score(X_Test,Y_Test)
from sklearn.ensemble import RandomForestRegressor
model7 = RandomForestRegressor(max_depth=1000)
model7.fit(X_Train, Y_Train)
```

```python
model7.score(X_Test,Y_Test)
```

```python
with open ('cropPrice.pkl','wb')as file:
    pickle.dump(model7,file)
```
