```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pickle
```

```python
data = pd.read_csv(r"../dataset/Crop_recommendation.csv")
data
```

```python
labelEncode = LabelEncoder()
data['label'] = labelEncode.fit_transform(data['label'])
data
```

```python
X = data.drop('label',axis=1)
Y = data['label']

X,Y
```

```python
# scaler = StandardScaler()

# X = scaler.fit_transform(X)
# X[0]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,shuffle=True,random_state=10)
X_test,Y_test
```

```python
model = RandomForestRegressor(max_depth=1000,min_samples_leaf=3)
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
labelEncode.classes_
```

```python
with open('cropPred.pkl','wb') as file:
    pickle.dump(model,file)
```
